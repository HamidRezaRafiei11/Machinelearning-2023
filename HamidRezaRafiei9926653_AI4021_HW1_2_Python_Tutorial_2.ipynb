**Problem 10:** Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:
* If 'Score' >= 90, Grade is 'A'
* If 'Score' >= 80, Grade is 'B'
* If 'Score' >= 70, Grade is 'C'
* If 'Score' < 70, Grade is 'D'

````python
import pandas as pd

data = {
    'Name' : ['x' , 'y' , 'z' , 'm' , 'n' , 'i' , 'v' , 'u' ,'o' , 'p'],
    'Score': [60 , 78 , 89 , 76 , 94 , 98 , 84 , 50 , 46 , 73]
}
df = pd.DataFrame(data)

def calculate_grade(score):
    if score >=90:
      return 'A'
    elif score >=80:
      return 'B'
    elif score >=70:
      return 'C'
    else:
      return 'D'


df['Grade'] = df['Score'].apply(calculate_grade)

print(df)

````

````python
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)


correlation = sales_df['Product A Sales'].corr(sales_df['Product B Sales'])

print(f'correlation between Product A and Product B sales : {correlation}')

if abs(correlation) >= 0.7:
    print("There is a strong correlation between Product A and Product B sales.")
else:
    print("There is no strong correlation between Product A and Product B sales.")
````

correlation:In data science and statistics, "correlation" refers to the statistical measure of the degree to which two or more variables or data sets are related to each other. Correlation provides insights into the strength and direction of the relationship between variables, and it is commonly used for various purposes in data analysis and modeling. Here are some key points about correlation in data science:

1. **Strength of Relationship:** Correlation quantifies how strongly two variables are related. It is expressed as a numeric value, often between -1 and 1.
   - A positive correlation (value close to 1) indicates that as one variable increases, the other tends to increase as well.
   - A negative correlation (value close to -1) indicates that as one variable increases, the other tends to decrease.
   - A correlation value close to 0 suggests a weak or no linear relationship between the variables.

2. **Correlation Coefficients:** There are several types of correlation coefficients used in data science, with Pearson correlation being the most common. Other correlation coefficients include Spearman's rank correlation and Kendall's tau. Each of these coefficients is suitable for different types of data and relationships.

3. **Scatterplots:** Correlation is often visualized using scatterplots, where data points are plotted on a graph to see how they are distributed. For a positive correlation, data points tend to form an upward-sloping pattern, while for a negative correlation, they form a downward-sloping pattern.

4. **Use Cases:** Correlation analysis is used in various data science tasks, including:
   - Feature selection: Identifying variables that are highly correlated with the target variable or with each other to avoid multicollinearity in regression models.
   - Data exploration: Understanding relationships between variables and identifying potential patterns.
   - Hypothesis testing: Evaluating the significance of relationships between variables.
   - Recommender systems: Identifying similarities or dissimilarities between users and items.
   - Financial modeling: Analyzing the relationships between financial variables.

5. **Limitations:** Correlation measures linear relationships but may not capture complex, non-linear relationships. Additionally, correlation does not imply causation, meaning that a strong correlation between two variables does not necessarily mean that one causes the other.

Correlation is a fundamental tool in data science for understanding and quantifying relationships between variables, making it an essential part of exploratory data analysis and predictive modeling.
